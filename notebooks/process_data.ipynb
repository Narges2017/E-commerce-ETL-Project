```python
from pyspark.sql.functions import to_date,col,trim,lower,when,round
```

```python
spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")
```

```python
df_transactions =spark.read.table("bronze_transactions")
```

```python
display(df_transactions.limit(5))
```

```python
spark.conf.set("spark.legacy.timeParserPolicy","LEGACY")
```

```python
df_transacitons = df_transacitons.withColumn("Date",to_date(col("Date"),"MM/dd/yyyy"))
```

```python
display(df_transacitons.limit(5))
```

```python
df_transacitons = df_transacitons.withColumn("ReturnFlag",when(col("Quantity")<1,True).otherwise(False))
display(df_transacitons.limit(3))
```

```python
df_transacitons = df_transacitons.withColumn("Revenu",round((col("Price")*col("Quantity")),2))
display(df_transacitons.limit(3))
```

```python
display(df_transacitons.filter(col("ProductNo").isNull()))
```

```python
df_transacitons = df_transacitons.filter(col("ProductNo").isNotNull())
```

```python
display(df_transacitons.limit(3))
```

```python
display(df_transacitons.filter(col("CustomerNo").isNull()))
```

```python
df_transacitons.write.format("delta").saveAsTable("silver_transactions")
```
